```python
import numpy as np
from sklearn.impute import KNNImputer

# Assuming you have a DataFrame 'df' with missing values

# Initialize the KNNImputer with the desired number of neighbors (k)
knn_imputer = KNNImputer(n_neighbors=5)  # You can adjust 'n_neighbors' as needed
data1 =  pd.DataFrame(data.Quantity)
# Fit and transform the DataFrame to perform k-NN imputation
data_imputed = knn_imputer.fit_transform(data1)

# The result is a NumPy array, so you can convert it back to a DataFrame if desired
df_imputed = pd.DataFrame(data_imputed, columns=data1.columns)
data.Quantity = df_imputed
data
# Now, 'df_imputed' contains the DataFrame with missing values imputed using k-NN
```
